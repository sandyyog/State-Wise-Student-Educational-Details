```python
df=df.rename({'No. of Inmates benefitted by Elementary Education':'IEE','No. of Inmates benefitted by Adult Education':'IAE','No. of Inmates benefitted by Higher Education':'IHE','No. of Inmates benefitted by Computer Course':'ICC'},axis=1)
df
```
